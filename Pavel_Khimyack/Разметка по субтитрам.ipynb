```python
class VideoDataset():
  df_files = None
  df_frames = None


  def __init__(self, df_files, df_frames, dataset_dir):
    self.df_files = df_files
    self.df_frames = df_frames
    self.file_id_to_name = self.df_files.set_index('file_id')['file_name'].to_dict()

    self.dataset_dir = dataset_dir
    self.video_dir = os.path.join(self.dataset_dir, "videos")

  def get_files(self):
    return self.df_files["file_name"].values

  def get_file_data_by_id(self, file_id):
    return self.df_files[self.df_files['file_id'] == file_id].to_dict(orient="records")[0]

  def file_name_by_id(self, file_id, fool_path = True):
    file_name = self.file_id_to_name.get(file_id, None)
    if fool_path:
      return os.path.join(self.video_dir, file_name)
    else:
      return file_name

```

```python
import pysrt
import datetime

class Phases(enum.Enum):
  STANCE = 0
  SWING = 1
  HIT = 2
  RETURN = 3

  RET = 4
  KICK = 5

def resolve_nearest_frame_no_by_milliseconds(milliseconds, fps:int):
  '''
  расчитывает номер ближайшего кадра по времени
  milliseconds - время в миллисекундах
  fps - количество кадров в секунду
  '''

  return round(fps * milliseconds  / 1000)

def resolve_start_end_frame_from_sub(sub: pysrt.SubRipItem, fps:int)->tuple:
  '''
  sub - элемент субтитра
  fps - количество кадров в секунду
  '''

  start_ms = sub.start.seconds * 1000 + sub.start.milliseconds
  end_ms = sub.end.seconds * 1000 + sub.end.milliseconds
  return [
      resolve_nearest_frame_no_by_milliseconds(start_ms, fps),
      resolve_nearest_frame_no_by_milliseconds(end_ms, fps)
  ]


def array_of_tupples_to_str_dot_comma(start_end_ms_tuple)->str:
  '''
    start_end_ms_tuple - кортеж вида (int, int)
    Преобразовать массив пар [[start, end] ...] -> start-end;....
  '''
  return  ";".join([ f"{start_frame_no}-{end_frame_no}" for start_frame_no, end_frame_no in  start_end_ms_tuple])

def resolve_phases_from_text(sub_text, phase:Phases):
  if sub_text in [phase.name.lower(),str(phase.value)]:
    #print(f"{sub_text} {True}")
    return True
  else:
    #print(f"{sub_text} {False}")
    return False


def parse_sub_file(sub_file_name):
  states_info = {
      Phases.STANCE.value: [],
      Phases.SWING.value: [],
      Phases.HIT.value: [],
      Phases.RETURN.value: [],
  }

  try:
      subs = pysrt.open(sub_file_name, encoding='utf-8')
      for i, sub in enumerate(subs):
          #print(f"Subtitle {i + 1}:")
          #print(f"Start time: {sub.start.seconds} {sub.start.milliseconds}")
          #print(f"End time: {sub.end} {sub.end.milliseconds}")

          phase_name = sub.text.strip()

          if resolve_phases_from_text(phase_name, Phases.STANCE):
              states_info[Phases.STANCE.value].append(resolve_start_end_frame_from_sub(sub, video_fps))

          elif  resolve_phases_from_text(phase_name, Phases.HIT) | resolve_phases_from_text(phase_name, Phases.KICK):
                states_info[Phases.HIT.value].append(resolve_start_end_frame_from_sub(sub, video_fps))

          elif  resolve_phases_from_text(phase_name, Phases.SWING):
              states_info[Phases.SWING.value].append(resolve_start_end_frame_from_sub(sub, video_fps))

          elif  resolve_phases_from_text(phase_name, Phases.RET) | resolve_phases_from_text(phase_name, Phases.RETURN) :
              states_info[Phases.RETURN.value].append(resolve_start_end_frame_from_sub(sub, video_fps))
          else:
              raise Exception(f"{phase_name} not supported")
  except Exception as e:
      print(f"An error occurred: {e}")
  return states_info
```

```python
df_for_process = mavashi_dzuki_all_df
result_dir = os.path.join(CFG_RESULT_EXCEL_DIR, Hits.MAVASHI_DZUKI.name)
xlsx_file_name = os.path.join(result_dir, "all.xlsx")


col_filename = "file"
col_stance = "stance(позиция)"
col_swing = "swing(замах)"
col_hit = "kick(удар)"
col_return = "return(возврат)"

states_infos = []

for idx, row in df_for_process.iterrows():
  file_name , video_fps, camera = row
  video_file_name = os.path.join(video_dataset.video_dir, file_name)
  sub_file_name = os.path.splitext(video_file_name)[0] + ".srt"
  base_file_name = os.path.basename(os.path.splitext(video_file_name)[0])
  print(base_file_name, camera)
  print(sub_file_name)
  states_info =  parse_sub_file(sub_file_name)
  states_info[col_filename] = f"{file_name}"
  states_infos.append(states_info)

print(f"создаем таблицу для {camera}")
data = {
    col_filename : [],
    col_stance : [],
    col_swing : [],
    col_hit : [],
    col_return : [],
}
print(states_infos)
for states_info in states_infos:
  data[col_filename].append(states_info[col_filename])
  data[col_stance].append(
      array_of_tupples_to_str_dot_comma(states_info[Phases.STANCE.value])
  )
  data[col_swing].append(
      array_of_tupples_to_str_dot_comma(states_info[Phases.SWING.value])
  )
  data[col_hit].append(
      array_of_tupples_to_str_dot_comma(states_info[Phases.HIT.value])
  )
  data[col_return].append(
      array_of_tupples_to_str_dot_comma(states_info[Phases.RETURN.value])
  )
df = pd.DataFrame(data)


os.makedirs(result_dir, exist_ok=True )
print(f"запись файла {xlsx_file_name}")
df.to_excel(xlsx_file_name, index=False)
```
